## LaTeX Basics

### Introduction
LaTeX is a typesetting system that allows you to focus on your content instead of formatting - formatting is done separately from entry.

You tell LaTeX "what it is" not "how it looks."

### Structure of a document

| Term | Description | Example |
|-|-|-|
| Command | Control sequence which performs an action | `\section`|
| Preamble | Block of commands that define the type of document you are writing,  the language you are writing in, the packages you would like to use. Comes before \begin{document}| `\documentclass{article}`|
| Package | Enable you to create bibliographies, insert images and figures, and write formulas. | `\usepackage{hyperref}` |
| Environment | Block of code with specific behavior depending on its type | `\begin{}` & `\end{}` |
| Body | Content of document enclosed inside an environment | `\begin{document}` |

```{note}
* Comments: 
Use % to create a comment. Nothing on the line after the % will be typeset.

* Restricted Characters: 
Certain symbols require a backslash to appear, like $, &, #, and %.
```

### Lists

Use the `\begin{itemize}...\end{itemize}` environment to create unnumbered lists, useful for detailing job responsibilities or committee service.

````{card}

```
\begin{itemize}
\item Apples
\item Cherries
\item Oranges
\item Peaches
\item Watermelon
\end{itemize}
```
---

This code results in:

- Apples
- Cherries
- Oranges
- Peaches
- Watermelon

````

Use the similar `\begin{enumerate}...\end{enumerate}` environment to create numbered lists.


## Publications Lists and Bibliographies in LaTeX

*Objective: Learn the basic commands to create and edit a list of publications and presentations within a bibliography file*

### Getting started with a .bib file

One way to manage publications, presentations and other outputs, is to reference a .bib file which is also used to 
create in-text citations and bibliographies. <br>
There are three ways to include a .bib file in a project in Overleaf.

1. Upload your own .bib file that you create or export from a citation manager.
2. Link to a URL (.bib)
3. Connect your Overleaf account with Mendeley or Zotero. 

### What does a citation in a .bib file look like? 

![bibkey](images/cv_bibentry.png)

The top line following @ is the “key” which is the syntax you use to add the citation to your document.
Your citation is only as good as its underlying data, so if something is misspelled or if you are missing a field in the .bib entry, errors will propagate. For example, if you would like to add a doi or url, you’ll need to add to the .bib entry in order for it to appear in the citation.

Find more information on formats: https://www.bibtex.com/format/


## Bibliography Packages

We will use the **biblatex** package to generate in-text citations and bibliographies. **Biblatex** is a flexible package for generating citations.

When adding a bibligraphy, or in this case, a list of publications and presentations, you will need to add commands to the preamble. 

Commands required for the **preamble**:

`\usepackage[backend=biber,style=apa]{biblatex}`

The above syntax calls in the biblatex package and the backend (biber), which acts as the interface between the .bib file and the LaTeX document. "Apa" denotes the citation style. This can be changed accordingly.

`\addbibresource{example.bib}`

The above command calls in the .bib file, which has the citation information for in-text citations and the bibliography.

In an article or book chapter, the command `\printbibliography` inserts the bibliography, which will contain citations referenced in the text. We will take a slightly different approach in adding references to a CV.

Find more information here: https://www.overleaf.com/learn/latex/Bibliography_management_in_LaTeX




### Exercise

````{card}
**Exercise 3: Adding a Bibliography**
^^^

*Objective: Learn to create, edit or upload a .bib file, use basic citation commands, and display a bibliography.*

**Create a .bib file and add references**
- Create a new file within your Overleaf project (click on the paper icon in the upper left) and name it CV_references.bib<br>
```{image} ./images/references.png
:alt: Google Scholar screenshot
:width: 400px
:align: center
``` 
- Search for this article in Google Scholar and locate the BibTeX formatted citation.<br>
   - 10.1080/15420353.2022.2080790
   
You can type a .bib entry from scratch, but it’s much easier to export the details from a database. 
Another option - if you have a Google Scholar profile, you can export your publications list as a bibtex file and then upload it or paste the contents into your Overleaf project.

```{image} ./images/scholar_bibtex.png
:alt: Google Scholar screenshot
:width: 400px
:align: center
``` 

- Paste the citation within your .bib file. (No preamble is needed).
Generally .bib records are findable in Google Scholar or other article databases.
<br>

**Adding a bibliography**
- To display bibliography in APA style, add package and style command to preamble:<br>
```
\usepackage[backend=biber,style=apa]{biblatex}
\addbibresource{CV_references.bib}
```

**Citation commands**<br>
**Publications**

Use the following commands to insert a publication into your CV.<br>
```
\section{Publications}
\subsection{2020-2021}
\begin{verse}
\fullcite{teplitzky2021introducing}
\end{verse}
```

**Presentations**

Presentations are less likely to have associated .bib entries. You will need to create your own entry in your .bib file. Use this syntax as a model:<br>
```
@presentation{agu2023,
    author={\textbf{Last name, first name} and Author2Last, Author2First and Author3Last, Author3First},
    title={Title of your presentation: use {Curly Brackets} to capitalize within title},
    eventtitle={AGU Fall Meeting 2023},
    venue={San Francisco, CA},
    eventdate={2023-12-12},
    url={www.conferenceurl.com},
}

```

Then use the following commands to insert a reference to a presentation into your CV.<br>
```
\section{Presentations}
\subsection{2020-present}
\begin{verse}
\fullcite{agu2023}
\end{verse}
```

**Tips**: Use subsections to group works by theme or date. Change font color or highlight text to identify particular articles or presentations.

<em>Commands needed:</em> `\fullcite{}`, `\section`, `\subsection`, `\begin{verse}...\end{verse}`

For additional examples and more information, please visit Overleaf's page on [bibliography management in LaTeX](https://www.overleaf.com/learn/latex/Bibliography_management_in_LaTeX)

````

## Downloading your CV

Once you've made your final edits, download a copy of your CV as a PDF.

![download](images/downloadpdf.png)

## Help and Troubleshooting

* https://www.bibtex.com/format/
* https://github.com/plk/biblatex-apa/blob/master/bibtex/bib/biblatex-apa-test-references.bib
* https://ctan.math.illinois.edu/macros/latex/contrib/biblatex/doc/biblatex.pdf
* https://www.overleaf.com/learn/how-to/Using_bibliographies_on_Overleaf
